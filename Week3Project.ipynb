```python
class ROICalc():

    def FullIncome(self):
        RentalIncome = int(input("How much is the rental income?: "))
        LaundryCost = int(input("How much is the laundry cost?: "))
        StorageCost = int(input("How much is the storage cost?: "))
        MiscCost = int(input("How much for other misc costs?: "))
        TotalIncome = RentalIncome+LaundryCost+StorageCost+MiscCost

        print(f"##########\nYour total income is ${TotalIncome}.\n##########")

    def FullExpenses(self):
        Tax = int(input("How much do you pay in taxes?. "))
        InsuranceCost = int(input("How much is the insurance cost?: "))
        UtilitiesCost = int(input("How much for utilities?: "))
        HOACost = int(input("How much is the HOA cost?: "))
        LawnSnowCost = int(input("How much is lawn care and snow removal?: "))
        VacancyCost = int(input("How much is the vacancy cost?: "))
        CopExCost = int(input("How much is the CopEx cost?: "))
        MortgageCost = int(input("How much is the mortgage cost?: "))
        Expenses = Tax+InsuranceCost+UtilitiesCost+HOACost+LawnSnowCost+VacancyCost+CopExCost+MortgageCost

        print(f"##########\nYour total expense cost is ${Expenses}.\n##########")

    def FullCashFlow(self):
        CashFlow = TotalIncome-Expenses
        print(f"##########\nYour Cash Flow is ${CashFlow}.\n##########")

    def FullInvestment(self):
        DownPayment = int(input("How much is the down payment?: "))
        ClosingCost = int(input("How much is the closing cost?: "))
        RehabBudget = int(input("How much is the rehab budget?: "))
        MiscOther = int(input("How much for other misc costs?: "))
        TotalInvestment = DownPayment+ClosingCost+RehabBudget+MiscOther
        print(f"##########\nYour total investment is ${TotalInvestment}.\n##########")

    def ROI(self):
        CashOnCashROI = (CashFlow*12)/TotalInvestment
        print(f"##########\nYour Cash on Cash ROI is {CashOnCashROI*100}%.\n##########")
        
def main():
    roicalc = ROICalc()
    while (response:=input("Would you like to calculate your ROI? You will need to know: \n1. Rental income, laundry cost, storage cost, other misc cost \n2. Tax, insurance, utilities, HOA cost, vacancy, CopEx, mortgage \n3. Down payment, closing costs, rehab budget, and misc other costs\n\nYou may input zero for any of the above numbers when prompted.\nReady? (Yes/No) ")):
        if response.lower() == "yes":
            roicalc.FullIncome()
            roicalc.FullExpenses()
            roicalc.FullCashFlow()
            roicalc.FullInvestment()
            roicalc.ROI()
            break
        if response.lower() == "no":
            break
main()
```
